# Desenvolver um programa que leia os elementos para um array [2][3] do tipo inteiro.
Após a entrada dos elementos calcular e exibir:

* 1.1. a soma  de todos os elementos;
* 1.2. a média de todos os elementos;
* 1.3. o maior elemento;
* 1.4. o menor elemento;
* 1.5. quantos são pares;
* 1.6. quantos são ímpares;
* 1.7. o elemento elevado ao quadrado;
* 1.8. porcentagem de elementos pares;
* 1.9. porcentagem de elementos ímpares;

```python
# Desenvolver um programa que leia os elementos para um array [2][3] do tipo inteiro.
# SENAI - Suiço Brasileiro.
# Brennon Meireles 1 DM - 1° semestre.
# 05/10/2023

import numpy as np

# variaveis zero para armazenar valores futuros:
soma = 0
maior = 0
menor = 0
par = 0
impar = 0

array = [[0,0,0],[0,0,0]] # lista de array

for i in range(2):      # criação de um for para cadastrar o valor da linha
  for j in range(3):    # criação de um for para cadastrar o valor da coluna
    array[i][j] = int(input(f'Digite o valor dos números [{i}][{j}] : '))  # input para guardar os valores dos n° digitados
    soma += array[i][j] # var para add e calcular a soma de todos os valores

# maior menor (dentro do for acima):

    # maior
    if  (maior < array[i][j]):
        maior = array[i][j]
    # menor
    if  (menor > array[i][j]):
        menor = array[i][j]

#  if para decidir qual número é impar ou par:
    if array[i][j] % 2 == 0:
        par += 1
    else:
      impar += 1

# porcentagem de impares e pares:
ppar = par / 6 * 100
pimpar = impar / 6 * 100



# Resultados:
print(f'A soma dos valores = {soma}')
print(f'A média dos valores = {soma/6}')
print(f'Número maior = {maior}')
print(f'Número menor = {menor}')
print(f'Quantidade de números pares = {par}')
print(f'Quantidade de números impares = {impar}')
print(f'A porcentagem de números impares = {pimpar} %')
print(f'A porcentagem números pares = {ppar} %')
```

```python
# versão resumida:

# Nesta versão, fiz as seguintes otimizações:
# Inicializei maior com um valor negativo e menor com um valor positivo para garantir que eles sejam atualizados corretamente.
# Usei np.zeros para criar uma matriz inicializada com zeros em vez de criar uma matriz vazia e preenchê-la depois.
# Usei operadores de comparação direta para encontrar o maior e o menor número, o que simplifica o código.
# Usei formatação de string para exibir as porcentagens com duas casas decimais.

import numpy as np

# variáveis para armazenar valores futuros:
soma = 0
maior = float('-inf')  # Inicializado com um valor negativo para encontrar o maior número corretamente
menor = float('inf')   # Inicializado com um valor positivo para encontrar o menor número corretamente
par = 0
impar = 0

array = np.zeros((2, 3), dtype=int)  # Crie uma matriz de zeros para armazenar os valores

for i in range(2):
    for j in range(3):
        array[i][j] = int(input(f'Digite o valor dos números [{i}][{j}] : '))
        soma += array[i][j]

        if array[i][j] > maior:
            maior = array[i][j]

        if array[i][j] < menor:
            menor = array[i][j]

        if array[i][j] % 2 == 0:
            par += 1
        else:
            impar += 1

# porcentagem de ímpares e pares:
ppar = (par / 6) * 100
pimpar = (impar / 6) * 100

# Resultados:
print(f'A soma dos valores = {soma}')
print(f'A média dos valores = {soma/6}')
print(f'Número maior = {maior}')
print(f'Número menor = {menor}')
print(f'Quantidade de números pares = {par}')
print(f'Quantidade de números ímpares = {impar}')
print(f'A porcentagem de números ímpares = {pimpar:.2f} %')
print(f'A porcentagem números pares = {ppar:.2f} %')
```
